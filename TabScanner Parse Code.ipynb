```python
import json
import pandas as pd
from pandas.io.json import json_normalize
import glob, os
```

```python
column_names = ["address","dateISO","establishment","lineItems","paymentMethod","summaryItems","taxes","total","filename"]
final = pd.DataFrame(columns = column_names)
```

```python
#os.chdir("Tabscanner Results")
for file in glob.glob("*.json"):
    filename = file
    f = open(filename,) 
    data = json.load(f) 
    
    new = pd.DataFrame.from_dict(data)
    new.reset_index(inplace = True) 
    new.drop('success', axis=1, inplace=True)
    new.drop('code', axis=1, inplace=True)
    cols = ['address','dateISO','establishment','lineItems','paymentMethod','summaryItems','taxes','total']
    new = new.loc[new['index'].isin(cols)]
    
    df_n = new.transpose()
    new_header = df_n.iloc[0] #grab the first row for the header
    df_n = df_n[1:] #take the data less the header row
    df_n.columns = new_header #set the header row as the df header
    df_n = df_n[3:]
    df_n['filename'] = filename
    
    final = final.append(df_n)
```

```python
final.reset_index(inplace = True)
#final
lineitems = final['lineItems'].apply(pd.Series)
#lineitems
final_merged = final.merge(lineitems, how='outer', left_index=True, right_index=True)
#final_merged
```

```python
for i in range(0,15):
    item1=final_merged[i].apply(pd.Series)
    item1 = item1[['desc','descClean','lineTotal','qty','productCode','unit']]
    itemdesc = 'desc_'+str(i)
    itemdescclean = 'descClean_'+str(i)
    itemlinetot = 'lineTotal_'+str(i)
    itemqty = 'qty_'+str(i)
    itemprodcode = 'productCode_'+str(i)
    itemunit = 'unit_' + str(i)
    
    final_merged[itemdesc]=item1['desc']
    final_merged[itemdescclean]=item1['descClean']
    final_merged[itemlinetot]=item1['lineTotal']
    final_merged[itemqty]=item1['qty']
    final_merged[itemprodcode]=item1['productCode']
    final_merged[itemunit]=item1['unit']
    
    #values = [item1['desc'], item1['descClean'], item1['lineTotal'],item1['qty'],item1['productCode'],item1['unit']]
    #final_merged2 = final_merged.merge(item1, how='outer', left_index=True, right_index=True)
```
